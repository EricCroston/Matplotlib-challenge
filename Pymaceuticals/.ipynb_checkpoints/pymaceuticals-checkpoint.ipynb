```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study = pd.merge(mouse_metadata, study_results, on=["Mouse ID"], how="left")

# Display the data table for preview
mouse_study.head()
```

```python
# Checking the number of mice.
mouse_count = len(mouse_study["Mouse ID"].unique())
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_timepoint = mouse_study[mouse_study.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_timepoint.head(10)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
g989 = mouse_study.loc[mouse_study['Mouse ID'] == "g989"]
g989
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_study_clean = mouse_study.drop(mouse_study[mouse_study['Mouse ID'] == "g989"].index)
mouse_study_clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count = len(mouse_study_clean["Mouse ID"].unique())
mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_vol = mouse_study_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median_vol = mouse_study_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
vol_var = mouse_study_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
vol_std = mouse_study_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = mouse_study_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary = pd.DataFrame({"Mean Tumor Volume":mean_vol, "Median Tumor Volume":median_vol, "Tumor Volume Variance":vol_var,
                        "Tumor Volume Std. Dev.":vol_std, "Tumor Volume Std. Err.":sem})
summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_sum = mouse_study_clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
agg_sum
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Put treatments into a list for for loop (and later for plot labels)
desired_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
selected_treatments = mouse_study_clean.loc[mouse_study_clean['Drug Regimen'].isin(desired_regimens)]

# Start by getting the last (greatest) timepoint for each mouse
mouse_max_timepoint = selected_treatments.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
promising_treatments = pd.merge(mouse_max_timepoint, mouse_study_clean, on=["Mouse ID", "Timepoint"], how="inner")
promising_treatments
```

```python

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []
regimen_tumor = pd.DataFrame()

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimen in desired_regimens:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    regimen_treatment = promising_treatments.loc[(promising_treatments['Drug Regimen'] == regimen)]
    tumor_vol.extend(regimen_treatment['Tumor Volume (mm3)'].tolist())

    # add subset 
    regimen_tumor[regimen] = tumor_vol
    
    # Determine outliers using upper and lower bounds
    quartiles = regimen_tumor[regimen].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = regimen_tumor.loc[(regimen_tumor[regimen] < lower_bound) | (regimen_tumor[regimen] > upper_bound), regimen].values
    print(f"{regimen}'s potential outliers: {outliers}")
    tumor_vol.clear()
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(regimen_tumor, 0, 'rs')
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()

```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin = mouse_study_clean[mouse_study_clean['Drug Regimen'] == 'Capomulin']
capomulin_mouse = capomulin[capomulin['Mouse ID'] == 's185']
capomulin_mouse_vol = capomulin_mouse[['Timepoint','Tumor Volume (mm3)']].set_index('Timepoint')
capomulin_mouse_vol.plot(kind="line", legend=False)
plt.title("Capomulin treatment of mouse S185")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin = mouse_study_clean[mouse_study_clean['Drug Regimen'] == 'Capomulin']
capomulin_mouse = capomulin.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().reset_index()
capomulin_mouse.columns = ['Mouse ID', 'Average Tumor Volume (mm3)']
capomulin_weight = capomulin.groupby('Mouse ID')['Weight (g)'].mean().reset_index()
capomulin_avg_vol = pd.merge(capomulin_mouse, capomulin_weight, on=['Mouse ID'], how='outer')
capomulin_avg_vol.plot(kind="scatter", x='Weight (g)', y='Average Tumor Volume (mm3)')
```
